```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
Obesity_df = pd.read_csv("obesity_data.csv")
Obesity_df.head()
```

```python
shuffled_Obesity_df = Obesity_df.sample(frac=1)
shuffled_Obesity_df
```

```python
#Lets try to make the labels numerical by assigning numberical Labels Manually
#Lets assign the following Labels
# Normal weight    1
# Overweight       2
# Obese            3
# Underweight      4

# shuffled_Obesity_df.iloc[0]["ObesityCategory"]
for i in range(1000):
    if shuffled_Obesity_df.loc[i, "ObesityCategory"] == "Normal weight":
        shuffled_Obesity_df.loc[i, "ObesityCategory"] = 1
    elif shuffled_Obesity_df.loc[i, "ObesityCategory"] == "Overweight":
        shuffled_Obesity_df.loc[i, "ObesityCategory"] = 2
    elif shuffled_Obesity_df.loc[i, "ObesityCategory"] == "Obese":
        shuffled_Obesity_df.loc[i, "ObesityCategory"] = 3
    else:
        shuffled_Obesity_df.loc[i, "ObesityCategory"] = 4
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder


np.random.seed(27)

x = shuffled_Obesity_df.drop("ObesityCategory",axis=1)
y = shuffled_Obesity_df["ObesityCategory"]

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)

categorical_columns = ["Gender" , "PhysicalActivityLevel"]

one_hot = OneHotEncoder()

transformer = ColumnTransformer([("one_hot",
                                one_hot,
                                categorical_columns)],
                                remainder='passthrough')

transformed_X_train = transformer.fit_transform(x_train)
transformed_X_test = transformer.transform(x_test)

label_encoder = LabelEncoder()
transformed_y_train = label_encoder.fit_transform(y_train)
transformed_y_test = label_encoder.transform(y_test)

model = RandomForestClassifier(n_estimators=100)

model.fit(
    transformed_X_train,
    transformed_y_train
)
```

```python
shuffled_Obesity_df["PhysicalActivityLevel"] = shuffled_Obesity_df["PhysicalActivityLevel"].astype(object)
dummies = pd.get_dummies(data=shuffled_Obesity_df[["Gender" , "PhysicalActivityLevel","ObesityCategory"]],
                         dtype=float)
shuffled_Obesity_df["Gender_Female"] = dummies["Gender_Female"]
shuffled_Obesity_df["Gender_Male"] = dummies["Gender_Male"]
shuffled_Obesity_df["PhysicalActivityLevel_1"] = dummies["PhysicalActivityLevel_1"]
shuffled_Obesity_df["PhysicalActivityLevel_2"] = dummies["PhysicalActivityLevel_2"]
shuffled_Obesity_df["PhysicalActivityLevel_3"] = dummies["PhysicalActivityLevel_3"]
shuffled_Obesity_df["PhysicalActivityLevel_4"] = dummies["PhysicalActivityLevel_4"]
shuffled_Obesity_df["ObesityCategory_1"] = dummies["ObesityCategory_1"]
shuffled_Obesity_df["ObesityCategory_2"] = dummies["ObesityCategory_2"]
shuffled_Obesity_df["ObesityCategory_3"] = dummies["ObesityCategory_3"]
shuffled_Obesity_df["ObesityCategory_4"] = dummies["ObesityCategory_4"]
```

```python
shuffled_Obesity_df = shuffled_Obesity_df.drop("Gender",axis=1)
shuffled_Obesity_df = shuffled_Obesity_df.drop("PhysicalActivityLevel",axis=1)
shuffled_Obesity_df = shuffled_Obesity_df.drop("ObesityCategory",axis=1)
shuffled_Obesity_df
```

```python
x_new = shuffled_Obesity_df[["Age","Height","Weight","BMI","Gender_Female","Gender_Male","PhysicalActivityLevel_1","PhysicalActivityLevel_2","PhysicalActivityLevel_3","PhysicalActivityLevel_4"]]
x_new
```

```python
y_new = shuffled_Obesity_df[["ObesityCategory_1","ObesityCategory_2","ObesityCategory_3","ObesityCategory_4"]]
y_new
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder


np.random.seed(27)

x_new_train , x_new_test , y_new_train , y_new_test = train_test_split(x_new,y_new,test_size=0.2)



model = RandomForestClassifier(n_estimators=100)

model.fit(
    x_new_train,
    y_new_train
)
```

```python
np.random.seed(27)

model.score(x_new_test,y_new_test)
```

```python
np.random.seed(27)

y_new_preds = model.predict(x_new_test)
y_new_preds
```
